##Task 2

###Create a simple book inventory system using tuples. Each book in the inventory should be
represented as a tuple containing the following information:

ÔÇ∑ Title

ÔÇ∑ Author

ÔÇ∑ Publication Year

ÔÇ∑ ISBN (International Standard Book Number)

ÔÇ∑ Quantity in Stock

1. Define an empty list named book_inventory to store tuples representing books.
2. Implement a function add_book that takes input for a new book and adds it to the
book_inventory. The function should prompt the user for the book details (title, author,
publication year, ISBN, quantity) and create a tuple to represent the book. Append this
tuple to the book_inventory.
3. Implement a function search_book that takes the ISBN as input and searches for the book
in the inventory. If the book is found, print its details; otherwise, print a message indicating
that the book is not in the inventory.
4. Implement a function display_inventory that prints the details of all books in the inventory.
Each book's information should be printed on a new line.
5. Implement a function sell_book that takes the ISBN and quantity as input and updates the
stock quantity of the corresponding book in the inventory. If the quantity is greater than the
available stock, print a message indicating insufficient stock.
6. Test your inventory system by adding a few books, searching for them, displaying the
inventory, and selling books.

```python
bookInventory = []

def addBook():
    BookInfo = (
        input("Enter the title of the book: "),
        input("Enter the author of the book: "),
        input("Enter the publication year of the book: "),
        input("Enter the ISBN of the book: "),
        int(input("Enter the quantity of the book: "))
    )
    bookInventory.append(BookInfo)
    print("‚úÖ Book added successfully.")

def searchBook(isbn):
    for i in range(len(bookInventory)):
        if bookInventory[i][3] == isbn:
            return i
    print("‚ùå Book not found.")
    return -1

def displayInventory():
    if not bookInventory:
        print("üì¶ Inventory is empty.")
    else:
        print("\nüìö Current Inventory:")
        for i, book in enumerate(bookInventory, start=1):
            print(f"{i}. Title: {book[0]}, Author: {book[1]}, Year: {book[2]}, ISBN: {book[3]}, Quantity: {book[4]}")

def sellBook():
    inputISBN = input("Enter the ISBN of the book: ")
    inputQuantity = int(input("Enter the quantity to sell: "))
    index = searchBook(inputISBN)
    if index == -1:
        return
    currentQuantity = bookInventory[index][4]
    if currentQuantity >= inputQuantity:
        updatedBook = (
            bookInventory[index][0],
            bookInventory[index][1],
            bookInventory[index][2],
            bookInventory[index][3],
            currentQuantity - inputQuantity
        )
        bookInventory[index] = updatedBook
        print("‚úÖ Book sold successfully.")
    else:
        print("‚ö†Ô∏è Insufficient stock.")

def menu():
    while True:
        print("\n--- üìò BOOK INVENTORY MENU ---")
        print("0: ADD BOOK TO INVENTORY")
        print("1: SEARCH BOOK IN INVENTORY")
        print("2: DISPLAY INVENTORY")
        print("3: SELL BOOK")
        print("4: EXIT")
        try:
            inputOption = int(input("Enter your option: "))
        except ValueError:
            print("‚ö†Ô∏è Please enter a valid number.")
            continue

        if inputOption == 0:
            addBook()
        elif inputOption == 1:
            isbn = input("Enter the ISBN of the book: ")
            index = searchBook(isbn)
            if index != -1:
                book = bookInventory[index]
                print(f"üîç Found Book: Title: {book[0]}, Author: {book[1]}, Year: {book[2]}, ISBN: {book[3]}, Quantity: {book[4]}")
        elif inputOption == 2:
            displayInventory()
        elif inputOption == 3:
            sellBook()
        elif inputOption == 4:
            print("üëã Exiting program. Goodbye!")
            break
        else:
            print("‚ùå Invalid option. Please try again.")

# Start the program
menu()
```
