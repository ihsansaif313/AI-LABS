✅ 1. Sort a 2D-array along different axe

```python
import numpy as np

# Sample 2D array
arr = np.array([[12, 5, 7], [3, 8, 1]])

# Sort along the first axis (rows)
sorted_first_axis = np.sort(arr, axis=0)

# Sort along the last axis (columns)
sorted_last_axis = np.sort(arr, axis=-1)

# Sort the flattened array
sorted_flattened = np.sort(arr.flatten())

print("Original:\n", arr)
print("Sorted along first axis:\n", sorted_first_axis)
print("Sorted along last axis:\n", sorted_last_axis)
print("Sorted flattened:\n", sorted_flattened)
```

✅ 2. Create a 3x3 matrix with values from 2 to 10

```python
matrix = np.arange(2, 11).reshape(3, 3)
print("3x3 Matrix:\n", matrix)
```

✅ 3. Convert Centigrade to Fahrenheit

```python
# Centigrade values stored in NumPy array
centigrade = np.array([0, 20, 37, 100])

# Conversion formula
fahrenheit = (centigrade * 9/5) + 32
print("Fahrenheit values:\n", fahrenheit)
```

✅ 4. Difference between max and min along last axi

```python
# Sample array
arr2 = np.array([[10, 20, 30], [5, 15, 25]])

# Difference along last axis
diff = np.max(arr2, axis=-1) - np.min(arr2, axis=-1)
print("Difference between max and min along last axis:\n", diff)
```

# TASK 2
Create two separate data frame using pandas with the information given in below tables.
{548EBC68-537E-482C-907A-3169CE676182}.png
Perform the following tasks:
 *  Merge both the data frames
 *  Sort w.r.t. employee salary in descending order
 *  Display the employee details with maximum salary
 *  Find how much variance exists between salaries of employees
 *  Make a histogram of the above data wrt. Salary having 5 bins.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create the two dataframes
df_emp = pd.DataFrame({
    'Emp_ID': [100, 110, 150, 123, 145, 112],
    'Emp_Name': ['Margaret', 'Chris', 'Allen', 'Susan', 'Simpson', 'Davis']
})

df_salary = pd.DataFrame({
    'Emp_ID': [100, 110, 150, 123, 145, 112],
    'Salary': [43000, 23000, 34000, 41200, 24000, 33300]
})

# Step 2: Merge both dataframes on Emp_ID
merged_df = pd.merge(df_emp, df_salary, on='Emp_ID')

# Step 3: Sort by Salary in descending order
sorted_df = merged_df.sort_values(by='Salary', ascending=False)

# Step 4: Display employee details with maximum salary
max_salary_emp = sorted_df.iloc[0]

# Step 5: Calculate variance in salaries
salary_variance = merged_df['Salary'].var()

# Step 6: Plot histogram with 5 bins
plt.hist(merged_df['Salary'], bins=5, edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.show()

# Display results
print("Merged DataFrame:\n", merged_df)
print("\nSorted by Salary (Descending):\n", sorted_df)
print("\nEmployee with Maximum Salary:\n", max_salary_emp)
print("\nVariance in Salaries:", salary_variance)
```
