##Task 3

```python
number=[]
while True:
  n= int(input(("Enter value")))
  if n==0:
    break
  number.append(n)
print(number)

if number:
  total=sum(number)
  print("sum of numbers are:",total)
  averageNum=total/len(number)
  print("Average of numbers are: ",averageNum)
```
