```python
from collections import deque

# 🧱 Node class with cost tracking
class Node:
    def __init__(self, name, parent=None, actions=[], cost=0):
        self.name = name
        self.parent = parent
        self.actions = actions
        self.cost = cost  # Total cost from start node

# 🔁 Reset graph before each run
def reset_graph(graph):
    for node in graph.values():
        node.parent = None
        node.cost = 0

# 🚀 BFS with cost accumulation
def bfs_with_cost(graph, start, goal):
    visited = set()
    queue = deque([graph[start]])
    graph[start].cost = 0  # Start node has zero cost

    print(f"\nStarting BFS from {start} to {goal}")

    while queue:
        current_node = queue.popleft()
        print(f"Visiting node: {current_node.name}, Cost so far: {current_node.cost}")

        if current_node.name == goal:
            print(f"Goal {goal} found! Reconstructing path...")
            return reconstruct_path_with_cost(current_node)

        visited.add(current_node.name)

        for neighbor_name, weight in current_node.actions:
            if neighbor_name not in visited:
                neighbor_node = graph[neighbor_name]
                if neighbor_node.parent is None:  # Avoid overwriting earlier path
                    neighbor_node.parent = current_node
                    neighbor_node.cost = current_node.cost + weight
                    print(f"  Adding neighbor {neighbor_name} with cost {neighbor_node.cost}")
                    queue.append(neighbor_node)

    print("Goal not reachable.")
    return None

# 🧩 Reconstruct path with cost
def reconstruct_path_with_cost(node):
    path = []
    while node:
        path.append((node.name, node.cost))
        node = node.parent
    path.reverse()
    print("Final path:", path)
    return path

# 🌐 Weighted graph structure
graph = {
    'A': Node('A', None, [('B', 6), ('C', 9), ('E', -1)]),
    'B': Node('B', None, [('A', 6), ('D', 3), ('E', 4)]),
    'C': Node('C', None, [('A', 9), ('E', 2), ('G', 3)]),
    'D': Node('D', None, [('B', 3), ('E', 5), ('F', 7)]),
    'E': Node('E', None, [('B', 4), ('C', 2), ('D', 5), ('F', 6)]),
    'F': Node('F', None, [('D', 7), ('E', 6), ('G', 7)]),
    'G': Node('G', None, [('C', 3)])
}


# 🧪 Example run: A to G
reset_graph(graph)
path = bfs_with_cost(graph, 'A', 'G')
```
