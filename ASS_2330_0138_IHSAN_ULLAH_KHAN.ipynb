# Water Jug Problem:

The water Jug Problem is a problem where two jugs of water are given, say one is a 4-litre one, and the other is a 3-litre one, but none of the measuring markers is mentioned. There is a pump available to fill the jugs with water. How can you exactly pour 2 liters of water into a 4-litre jug? Assuming that both the jugs are empty, the task is to find a solution to pour 2-litre water into a 4-litre jug.

    Formulate the problem (Write its states, initial state, actions, transition model, goal test and path cost).

Solution:

States: A state can be represented as a tuple (x, y) where:

    x represents the amount of water in the 4-litre jug.

    y represents the amount of water in the 3-litre jug.

Initial State: The initial state is (0, 0), meaning both jugs are empty.

Actions: Possible actions to move from one state to another:

    Fill the 4-litre jug: (4, y)

    Fill the 3-litre jug: (x, 3)

    Empty the 4-litre jug: (0, y)

    Empty the 3-litre jug: (x, 0)

    Pour water from the 4-litre jug into the 3-litre jug until the 3-litre jug is full or the 4-litre jug is empty:

    If x + y <= 3: (0, x + y)

    If x + y > 3: (x - (3 - y), 3)

    Pour water from the 3-litre jug into the 4-litre jug until the 4-litre jug is full or the 3-litre jug is empty:

    If x + y <= 4: (x + y, 0)

    If x + y > 4: (4, y - (4 - x))

**Transition Model:** The transition model defines the result of applying an action to a state. For example, if you are in state (0, 0) and you fill the 4-litre jug, you transition to state (4, 0).

**Goal Test:** The goal is to have exactly 2 litres of water in the 4-litre jug. The goal state is (2, y) where y can be any value between 0 and 3.

**Path Cost:** Each action has a uniform cost, which can be considered as 1.
